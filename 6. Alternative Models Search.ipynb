# Creating traning/testing df

```python
import pandas as pd
```

## Importing Data:

```python
# importing and basic formating
train = pd.read_csv('C:/Users/danfv/OneDrive/Knowledge/Courses/Data Analysis/LHL - Data Analysis Bootcamp/lighthouse-data-notes/W6 Midterm Project/Midterm_Project/Data/Flights - Sample 10000 - Peter Dan Features.csv').sort_values('fl_date').reset_index(drop=True)
train = train.drop('Unnamed: 0', axis=1)

# deleting features the model won't use
date_and_carrier = ['fl_date', 'mkt_unique_carrier', 'branded_code_share', 'mkt_carrier', 'op_unique_carrier', 'tail_num', 'op_carrier_fl_num']
origin_and_dest = ['origin_airport_id', 'origin', 'origin_city_name', 'dest_airport_id', 'dest', 'dest_city_name']
drop_streamlining = ['tail_num', 'dup', 'flights']
drop_peter_string = ['dep_weather', 'arr_weather']
drop_redundant_feat = ['distance']

low_p_val_list = ['season', 'weekday', 'month', 'arr_taxi_ind', 'airtime_ind',
    'orig_airp_fl_ind', 'dest_airp_fl_ind', 'orig_airp_pss_ind', 'dest_airp_pss_ind',
    'origin_airport_freq', 'dest_airport_freq']

dropping = date_and_carrier + origin_and_dest + drop_streamlining + drop_peter_string + drop_redundant_feat + low_p_val_list
train = train.drop(dropping, axis=1)

# # Separating features and target for regression:
X = train.drop('arr_delay', axis=1)
y = train.arr_delay
```

```python
y
```

## Train/Test Split:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

# Running Lazy Predict:

```python
import lazypredict
```

```python
from lazypredict.Supervised import LazyRegressor

reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
models, predictions = reg.fit(X_train, X_test, y_train, y_test)

print(models)
```
