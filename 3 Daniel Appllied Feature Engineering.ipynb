```python
# flying season
# https://www.planestats.com/season_2015oct

def is_peak_season(month):
    '''
    Return a number indicating flying season

        Paraneters:
            x(int): month number 1-12
        
        Returns:
            int: 0 for low season, 1 for medium, 2 for high
    '''
    
    if month >= 1 and month <= 2:
        return 0
    elif month >= 3 and month <= 5:
        return 1
    elif month >= 6 and month <= 8:
        return 2
    elif month >= 9 and month <= 11:
        return 0
    else:
        return 1

is_peak_season(12)

flights['month'] = pd.DatetimeIndex(flights['fl_date']).month
flights['peak_season'] = flights['month'].apply(lambda x: is_peak_season(x))
```
